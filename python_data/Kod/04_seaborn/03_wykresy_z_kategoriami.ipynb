```python
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
tips.head()
```

STRIPPLOT

```python
_ = sns.catplot(tips, x='day', y='total_bill')
```

SWARMPLOT

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='swarm')
```

BOXPLOT

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='box')
```

BAR

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='bar')
#kresecki to przedzia to przedziały ufnosci (95%)
```

COUNTPLOT

```python
sns.catplot(tips, x='day', kind='count')
```
